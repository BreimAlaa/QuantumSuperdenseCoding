```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer, assemble
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram, array_to_latex

qc = QuantumCircuit(2)

state = Statevector.from_instruction(qc)
display(state.draw('bloch'))
```

```python
qc.h(1)
qc.barrier(label = 'H gate')
print('After H gate')
state = Statevector.from_instruction(qc)
display(state.draw('latex'))
display(state.draw('bloch')) 
```

```python
qc.cx(1,0)
state = Statevector.from_instruction(qc)
qc.barrier(label = 'CNOT gate')
print('After CNOT gate')
display(state.draw('latex'))
display(state.draw('bloch')) 
```

```python
qc.z(1)
state = Statevector.from_instruction(qc)
qc.barrier(label = 'Z gate')
print('After Z gate')
display(state.draw('latex'))
display(state.draw('bloch')) 
```

```python
qc.x(1)
state = Statevector.from_instruction(qc)
qc.barrier(label = 'X gate')
print('After X gate')
display(state.draw('latex'))
display(state.draw('bloch')) 
```
